```python
from os import name as os_name, system as terminal


def clear():
    terminal('clear' if os_name.lower() == 'nt' else 'clear')

def register_phone_number():
    clear()
    print("\n\n------ Register ------")
    phone_number = input("Enter phone number [valid phone number = 09123334444] : ")

    
    assert len(phone_number) == 11 and phone_number[:2] == '09', "phone number is not valid !"

    fullname = input("Enter phone name : ")


    assert not fullname.isnumeric(), "fullname not valid !"

    with open("phonebooks.txt", "a+") as db:
        db.write(f"{fullname}:{phone_number}\n")

    input("Your number saved ! \n\nPress enter to back main ... ")
```

```python
def search_phone_numbers():
    clear()
    print("\n\n------ Search ------")
    lines = []
    results = []
    
    
    menu_items = [
        'phone number',
        'phone name',
    ]
    
    for index, value in enumerate(menu_items):
        print(index + 1, value)
    
    try:
        user_plan = int(input("> "))

    except (ValueError, IndexError):
        input("Error !!! \ninvalid key! \n\nPress enter to back main ... ")
        
    assert user_plan == 1 or user_plan == 2, "user input is not valid ! \nPress enter to try again!"
    

    
    if user_plan == 2: 
        fullname = input("Enter phone name : ")
    
        fullname.replace(" ", "")
    else: 
        fullname = input("Enter phone number [valid phone number = 09123334444] : ")
        
        fullname.replace(" ", "")

    
    with open('phonebooks.txt') as f:
        while True:
            line = f.readline()
            line = line.replace("\n", "")
            line = line.replace(" ", "")
            if not line:
                break
            lines = line.split(":")
            if fullname in lines[0]:
                results.append(lines[1])
            if fullname in lines[1]:
                results.append(lines[0])
                        
    assert results, "user input is not valid ! \nPress enter to try again!"
    
    z = 0
    for i in results:
        z += 1
        print(z, ": ", i)
        
    input("\nPress enter to back main ... ")
```
